1) Выявим колонки с пропусками

```python
for col in df.columns:
    mis_val = np.mean(df[col].isnull())
    print(f'{col} - {round(mis_val * 100)}%')
```

```python
# Удалим колонки с неполными и лишними данными
df.drop(['Club','Value', 'Contract Valid Until', 'Overall', 'Potential', 'ID'], axis= 1 , inplace= True)
```

2) Найдём полные дубликаты и удалим их.

```python
df = df.drop_duplicates()
df.shape
```

```python
# 4 строки-дубликаты удалились
df.head()
```

3) Создадим колонку с разбиением возраста по группам

```python
df['Age'] = df['Age'].astype(int)

def new_column(col='Age'):
    
    my_list = []
    age_list = list(df[col])
    
    for i in age_list:
        if i < 20:
            my_list.append(1)
        elif i>20 and i<30:
            my_list.append(2)
        elif i>30 and i<36:
            my_list.append(3)
        else:
            my_list.append(4)
    df.insert (loc=2 , column='Age_group', value=my_list)
    return df.head()
```

```python
new_column(col='Age')
```

```python
print(len(df[df['Age_group']==1]))
print(len(df[df['Age_group']==2]))
print(len(df[df['Age_group']==3]))
print(len(df[df['Age_group']==4]))
```
